# Fine-tuning

```python
model = YOLO(r"/content/detection/yolo/weights/yolo26l.pt")

# Замораживаем первые 150 слоев
model.freeze = 150
```

```python
# Обучаем
results = model.train(
    data=r"/content/dataset/data.yaml",
    project=predict_path,
    name='yolo_finetuned',
    exist_ok=True,

    save=True,
    save_period=2,

    imgsz=1024,
    batch=8,
    nbs=32,

    epochs=10,  
    patience=7, 

    device=0,

    cos_lr=True,

    hsv_h=0.005,
    hsv_s=0.25,
    hsv_v=0.25,
    degrees=1.0,  
    translate=0.1,  
    scale=0.3,  
    fliplr=0.5,
    mosaic=0.5,  
    close_mosaic=5, 

    single_cls=True,
    seed=999,

    resume=False,  
    lr0=0.002,  
    warmup_epochs=2,  
)
```
